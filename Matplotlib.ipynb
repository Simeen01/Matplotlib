```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Pie Chart

```python
areas = ["Marketing","Sales","Developers","HR","Customer Support"]
budget = [3,4,10,1,2.5]
plt.pie(budget,labels=areas,autopct="%0.3f%%") #0.3 - 3 numbers after decimal place
plt.show() #removes all warning
```

Scatter plot

```python
x = [12,45,87,23,98]
y = [74,83,99,43,26]
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age wise Salary Comparison")
plt.scatter(x,y)
plt.show()
```

Line plot

```python
product_id = [1,2,3,4,5,6]
online_prices = [200,450,670,900,500,780]
offline_prices = [150,300,800,1000,200,500]
chor_bazar = [100,150,400,200,50,40]
plt.plot(product_id,online_prices,marker="D",color = "pink",label="Online Prices")
plt.plot(product_id,offline_prices,marker="*",color="black",label="Offline Prices")
#plt.plot(product_id,online_prices,offline_prices,marker = "D",markersize=10) #marker - points
plt.plot(product_id,chor_bazar,marker="+",color="blue",label="Chor Bazar")
plt.legend()
plt.rcParams["figure.figsize"]=(7,7) #plot size
plt.xlabel("Product")
plt.ylabel("Price")
plt.title("Product VS Prices")
plt.show()
```

Subplot

```python
plt.subplot(1,3,1) #(row,col,position)
plt.plot(product_id,online_prices,marker="D",color = "pink",label="Online Prices")

plt.subplot(1,3,2)
plt.plot(product_id,offline_prices,marker="*",color="black",label="Offline Prices")

plt.subplot(1,3,3)
plt.plot(product_id,chor_bazar,marker="+",color="blue",label="Chor Bazar")

plt.xlabel("Product")
plt.ylabel("Price")
plt.title("Product VS Prices")
plt.figure(figsize=(6,6))
```

Bar Graph

```python
year = [2012,2013,2014,2015,2016,2017,2018,2019,2020]
price = [70000,80000,88000,90000,60000,50000,40000,55000,95000]
plt.bar(year,price,color="purple")
#plt.rcParams["figure.figsize"]=(6,6)
plt.xlabel("Year")
plt.ylabel("Prices")
plt.title("Year VS Prices")
plt.figure(figsize=(5,4))
plt.show()
```
